```python
# For this exercise, DF_graph needs to be copied in (which is the two cells found underneath)
import pandas as pd

DF_edges = pd.DataFrame([['b', 'a','60'],
 ['a', 'c','50'],
 ['a', 'd','100'],
 ['a', 'e','80'],
 ['a', 'g','70'],
 ['b', 'c','20'],
 ['g', 'i','70'],
 ['g', 'c','40'],
 ['h', 'i','20'],
 ['i', 'd','60'],
 ['i', 'j','60'],
 ['d', 'e','5'],
 ['e', 'k','10'],
 ['f', 'l','50'],
 ['f', 'k','25'],
 ['j', 'k','80']],
columns=['node1', 'node2','weight'])

DF_edges["weight"] = DF_edges["weight"].astype(str).astype(float)

DF_edges
```

```python
#For this exercise, DF_graph needs to be copied in:
DF_graph=nx.Graph()
DF_graph = nx.from_pandas_edgelist(DF_edges, 'node1', 'node2', ['weight'])

pos = nx.spring_layout(DF_graph)
label_weight = nx.get_edge_attributes(DF_graph, "weight")
nx.draw(DF_graph,
        pos,
        with_labels = True,
        node_color='darkorchid',
        node_size=200,
        font_color="white",
        font_weight='bold',
        font_size=8)
nx.draw_networkx_edge_labels(DF_graph,pos=pos,edge_labels=label_weight,font_size=8)
plt.show()
```

```python
#Two nodes with the highest degrees
deg=nx.degree(DF_graph, weight="weight")
deg
```

```python
#A and I have the highest degrees. Now we have to find the people connecting A and I
calls_path=nx.shortest_path(DF_graph, source='a', target='i', weight='inv_weight')
print (calls_path)
call_edges=list(zip(calls_path, calls_path[1:]))
print(call_edges)
```
